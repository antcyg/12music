```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

data = pd.read_table('iqsize.txt')
```

```python
data.tail(10)
```

```python
for x_str in ['Brain', 'Height', 'Weight']: 
    x = sm.add_constant(data[x_str])
    model = sm.OLS(data['PIQ'],x).fit()
    print(model.pvalues)
    print('===============')
```

```python
for x_str in ['Height', 'Weight']: 
    x = sm.add_constant(data[['Brain',x_str]])
    model = sm.OLS(data['PIQ'],x).fit()
    print(model.pvalues)
    print('===============')
```

```python
x = sm.add_constant(data[['Brain','Height', 'Weight']])
model = sm.OLS(data['PIQ'],x).fit()
model.summary()
```
