财务概念：accrual accouting <br>
交易发生即被记录 <br>
现金进出才被记录 <br>
例：提前订票，付了钱，直到演唱会开始交易才发生

#### Operating Cashflow 营业现金流 业务相关 与购买资产设备 借钱还钱 无关 <br>
例如：波音 2015至今 net operating cash flow; 谷歌，苹果 <br>
与净利润进行比较，判断盈利质量 <br>
例如：亚马逊，营业现金流大幅领先盈利，本可以操纵盈利，反映出盈利质量高 <br>
例如：netflix，一度盈利增长，营业现金流下降，实际盈利能力堪忧 <br>
大部分企业，两者比例比较稳定：营业现金流除以净利润

#### Investing Cashflow：资产，关注是否投资自己
例如 亚马逊 <br>
如果为正，可能是在变卖资产，缓解压力 <br>

#### Finance：发股 发债 还债 回购 派息
例如：成熟企业 costco 麦当劳

#### 综合指标： free cash flow
理论上说 越高越好 <br>
但是要防止通过减少对自己投资 获得短期财报表现 <br>
例如 under armour; 2017年 dollar general，dollar tree <br>
